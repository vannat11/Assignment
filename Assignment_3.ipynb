```python
import re
import spacy
from spacy import displacy 
from spacy.tokens import DocBin
```

```python
NER = spacy.load('en_core_web_sm')
```

```python
# Download the training data
!wget https://groups.csail.mit.edu/sls/downloads/restaurant/restauranttrain.bio

```

```python
# Read the training data and preprocess it
with open('restauranttrain.bio') as f:
     text = f.read()
     text = re.sub(r'([A-Z])\n', r'\1 O\n', text)
     text = text.replace('-\n', '')
     text = text.strip()
```

```python
text
```

```python
# prepare an empty model to train
nlp = spacy.blank('en')
nlp.vocab.vectors.name = 'demo'
ner = nlp.create_pipe('ner')
nlp.add_pipe(ner, last = True)
```

```python
# Load the spaCy model
nlp = spacy.blank('en')
```

```python
# Convert the preprocessed training data to spaCy format


doc_bin = DocBin()
for line in text.split('\n\n'):
    tokens = line.split()
    words = [token.split('|')[0] for token in tokens]
    entities = [token.split('|')[1] for token in tokens]
    doc = nlp(' '.join(words))
    for i, token in enumerate(doc):
        if entities[i] != 'O':
            doc.ents = list(doc.ents) + [(i, i+1, entities[i])]
    doc_bin.add(doc)
doc_bin.to_disk('train.spacy')
```

```python
# Load the dataset into Spacy
nlp = spacy.blank("en")
ner = nlp.create_pipe("ner")
nlp.add_pipe(ner)
```
